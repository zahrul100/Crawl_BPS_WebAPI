# Melakukan Crawling WebAPI BPS

```python
import pandas as pd #import Library
```

```python
id = str(521) #sample id yang digunakan
```

```python
df = pd.read_json('https://sig.bps.go.id/rest/getdatadetail?id='+id) #melakukan read json file dengan pandas
```

```python
#mengambil informasi berupa kategori,judul dan berdasar
kategori = df['response']['symbol']['name']
judul = df['response']['long_title']
berdasar = df['response']['variable']
```

```python
print("kategori : "+kategori)
print("Judul : "+judul)
print("Berdasar :"+berdasar)
```

```python
series = df['response']['series']
```

```python
series
```

```python
#Menyimpan series_id untuk digunakan pada url
#from tqdm import tqdm
series_id = []
string_series_id = ''
for idx in range(len(series)):
#   print(series[idx]['series_id'])
    series_id.append(series[idx]['series_id'])
    if(idx == 0):
        string_series_id = str(series[idx]['series_id'])
    else:
        
        string_series_id = string_series_id+'%2C'+str(series[idx]['series_id'])
```

```python
new_url = 'https://sig.bps.go.id/rest/getdataseries?data_id='+str(id)+'&series_id=('+string_series_id+')'
```

```python
new_url
```

```python
df2 = pd.read_json(new_url) #membaca url baru
```

```python
df2
```

```python
df3 = pd.json_normalize(df2['response']['series']) #menormalisasikan json
```

```python
df3
```

```python
final = {'kode':[],'provinsi':[]} #membuat dataframe baru untuk menyimpan hasil akhir json
```

```python
final = pd.DataFrame(final)
```

```python
for idx in range(len(df3)): #menggabungkan semua series data
    nilai = pd.json_normalize(df3['series_data'][idx]) #menormalisasikan json
    final[str(df3['series_year'][idx])+'_'+df3['series_period_value'][idx]] = '' #membuat kolom tahun bulan
    final['kode'] = nilai['kode'] #mengisi kolom kode
    final['provinsi'] = nilai['label'] #mengisi kolom provinsi
    final[str(df3['series_year'][idx])+'_'+df3['series_period_value'][idx]] = nilai['value'] #mengisi kolom tahun bulan
    
```

```python
final
```

```python
namafile = kategori+'_'+judul+'_'+berdasar+'.csv'
```

```python
final.to_csv(namafile,sep=';') #menyimpan file dalam bentuk csv
```
