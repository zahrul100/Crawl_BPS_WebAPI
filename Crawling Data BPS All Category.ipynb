```python
#237-240,340,588,669,670,680,683 error
import pandas as pd
for data_id in range(1,700):
    df = pd.read_json('https://sig.bps.go.id/rest/getdatadetail?id='+str(data_id))
    if df.empty:
        print(str(data_id)+'--------------------------->Kosong')
        continue
    print(str(data_id)+'---------------------------------------------------->Progres')
    kategori = df['response']['symbol']['name']
    judul = df['response']['long_title']
    berdasar = df['response']['variable']
    kelompok = df['response']['subject_child_id']
    nama_kolom = judul+' '+berdasar
    karakter = '/'
    nama_kolom = nama_kolom.replace(karakter,'')
    karakter = ')'
    nama_kolom = nama_kolom.replace(karakter,'')
    karakter = '('
    nama_kolom = nama_kolom.replace(karakter,'')
    karakter = '*'
    nama_kolom = nama_kolom.replace(karakter,'')
    karakter = '!'
    nama_kolom = nama_kolom.replace(karakter,'')
    karakter = ';'
    nama_kolom = nama_kolom.replace(karakter,'')
    
    karakter = ':'
    nama_kolom = nama_kolom.replace(karakter,'')
    
    karakter = '>'
    nama_kolom = nama_kolom.replace(karakter,'Lebih Besar')
    karakter = '<'
    nama_kolom = nama_kolom.replace(karakter,'Kurang Dari')
    karakter = '='
    nama_kolom = nama_kolom.replace(karakter,'Sama Dengan')
    karakter = ' '
    nama_kolom = nama_kolom.replace(karakter,'_')

    
    
    print('-------------------------------------------------------->Kategori : '+kategori)
    print('------------------------------------------------------------>Judul : '+judul)
    print('------------------------------------------------------------>Berdasar : '+berdasar)
#     namafile = kategori+'_'+judul+'_'+berdasar
    
    
#     df['response']['series'][data_id]['series_id']
    series = df['response']['series']
    if series:
#         print('series ada')
        print(' ')
    else:
        continue
    series_id = []
    string_series_id = ''
    for idx in range(len(series)):
#             print(series[idx]['series_id'])
            
            series_id.append(series[idx]['series_id'])
            if(idx == 0):
                string_series_id = str(series[idx]['series_id'])
            else:

                string_series_id = string_series_id+'%2C'+str(series[idx]['series_id'])
    if len(series) > 1 :
        new_url = 'https://sig.bps.go.id/rest/getdataseries?data_id='+str(data_id)+'&series_id=('+string_series_id+')'
    else:
        new_url = 'https://sig.bps.go.id/rest/getdataseries?data_id='+str(data_id)+'&series_id=('+string_series_id+'%2C000)'
    
    df2 = pd.read_json(new_url)
    index = df2.index
    
    df3 = pd.json_normalize(df2['response']['series'])
    nilai = pd.json_normalize(df3['series_data'][0])
    final = {'kode':[],'provinsi':[]}
    final = pd.DataFrame(final)
    final['kode'] = nilai['kode']
    final['provinsi'] = nilai['label']
    final[str(df3['series_year'][0])+'_'+df3['series_period_value'][0]] = nilai['value']
    for idx in range(len(df3)):
        nilai = pd.json_normalize(df3['series_data'][idx])
        if nilai.empty:
            print('nilai array '+str(idx)+' kosong')
        else:
            final[nama_kolom+'_'+df3['series_period_value'][idx]+'_'+str(df3['series_year'][idx])] = ''
            final['kode'] = nilai['kode']
            final['provinsi'] = nilai['label']
            final[nama_kolom+'_'+df3['series_period_value'][idx]+'_'+str(df3['series_year'][idx])] = nilai['value']
            
    namafile = str(kelompok)+'+'+kategori+'+'+judul+'+'+berdasar+'.csv'
    bad_chars ='/'
    namafile = namafile.replace(bad_chars,'')
    bad_chars ='='
    namafile = namafile.replace(bad_chars,'Sama Dengan')
    bad_chars ='>'
    namafile = namafile.replace(bad_chars,'Lebih Dari')
    bad_chars ='<'
    namafile = namafile.replace(bad_chars,'Kurang Dari')
    namafile = namafile.replace("Rata-Rata",'Mean')
    
    final.to_csv('E:\Magang\crawl_bps/'+namafile,sep=';',index = False)
```
